```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
sns.histplot(df['math score'])
plt.title("Distribution of Math Score")
plt.show()
```

Description: This histogram shows the distribution of math scores, with most students scoring between 50 and 80. The peak around the mid-60s indicates that the majority achieved average performance, while extreme scores are less common.

```python
sns.histplot(df['reading score'])
plt.title("Reading Score Distribution")
plt.show()
```

Description: This histogram shows the distribution of reading scores. Most students scored between 60 and 85, with a peak around the low 70s, indicating generally strong performance in reading compared to lower score ranges.

```python
sns.histplot(df['writing score'])
plt.title("Writing Score Distribution")
plt.show()
```

Description: This histogram shows the distribution of writing scores. Most students scored between 60 and 85, with a peak around the low to mid-70s, indicating generally strong writing performance among the majority of students.

```python
df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
```

Description: This table compares the average scores based on test preparation course completion. Students who completed the course achieved higher average scores in math, reading, and writing compared to those who did not, indicating a positive impact of test preparation on performance.

```python
df.groupby('gender')[['math score','reading score','writing score']].mean().plot(kind='bar')
plt.title("Scores by Gender")
plt.show()
```

Description: This bar chart compares average scores by gender across math, reading, and writing. Females outperform males in reading and writing, while males have slightly higher average scores in math, showing a noticeable performance difference between genders across subjects.

```python
df.boxplot(column=["math score","reading score","writing score"])
plt.show()
```

Description: This boxplot compares the distribution of math, reading, and writing scores. Reading and writing have slightly higher medians than math, while all subjects show similar spread and a few low-score outliers, indicating generally consistent performance across subjects.

```python
plt.scatter(df["math score"], df["reading score"])
plt.xlabel("Math")
plt.ylabel("Reading")
plt.show()
```

Description: This scatter plot shows the relationship between math and reading scores. There is a strong positive correlation, as students with higher math scores tend to also have higher reading scores, indicating consistent performance across subjects.

```python
import seaborn as sns

sns.heatmap(df[['math score','reading score','writing score']].corr(), annot=True)
plt.title("Correlation Between Scores")
plt.show()
```

Description: This heatmap displays the correlation between math, reading, and writing scores. All subjects show strong positive correlations, with the highest correlation between reading and writing, indicating that students who perform well in one subject tend to perform well in the others.

# Insights:

There is a strong positive relationship between scores across all
subjects.

Students who perform well in one subject tend to perform well in others.

Reading and writing scores are generally higher than math scores.

Females perform better in reading and writing, while males slightly lead in math.

Completing the test preparation course is associated with higher scores in all subjects.

Students with better overall academic performance show consistent results across different visualizations.
